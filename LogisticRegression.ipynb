```python
import pandas as pd
```

```python
titanic_df = pd.read_csv('datasets/titnaic_processed.csv')
titanic_df.head()
```

```python
titanic_df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = titanic_df.drop('Survived', axis = 1)
Y = titanic_df['Survived']
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(penalty='l2', C = 1.0, solver='liblinear').fit(x_train,y_train)
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
pred_results = pd.DataFrame({'y_test': y_test, 'y_pred': y_pred})
```

```python
pred_results.head()
```

```python
titanic_cross_tab = pd.crosstab(pred_results.y_pred, pred_results.y_test)
titanic_cross_tab
```

```python
from sklean.metrics import acccuracy_score
from sklean.metrics import precision_score
from sklean.metrics import recall_score

acc = acccuracy_score(y_test, y_pred)
prec = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

print("Acc: ", acc)
print("prec: ", prec)
print("recall: ", recall)
```
